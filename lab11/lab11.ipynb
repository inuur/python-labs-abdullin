[4]. Создайте в DataFrame data новый признак weight_category, который будет иметь 3 значения: 1 – если вес меньше
120 фунтов. (~ 54 кг.), 3 - если вес больше или равен 150 фунтов (~68 кг.), 2 – в остальных случаях. Постройте «ящик с
усами» (boxplot), демонстрирующий зависимость роста от весовой категории. Используйте метод boxplot библиотеки
Seaborn и метод apply Pandas DataFrame. Подпишите ось y меткой «Рост», ось x – меткой «Весовая категория».

```python
def check(row):
    if row['Weight(Pounds)'] < 120:
        return 1
    elif row['Weight(Pounds)'] > 150:
        return 3
    else:
        return 2

df['weight_category'] = df.apply(lambda row: check(row), axis=1)
df.head()
```

```python
sns.boxplot(x="Weight(Pounds)", y="Height(Inches)", data=df)
```
